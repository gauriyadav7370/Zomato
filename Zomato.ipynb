Data Analysis Project ZOMATO

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

pandas is used for data manipulation and analysis.
numpy is used for numerical operations.
matplotlib.pyplot and seaborn are used for data visualization.

 Create the data frame.


```python
dataframe =  pd.read_csv("Zomato data .csv")
print(dataframe.head())
```

```python
dataframe
```

convert the data type of the "rate" column to float and remove the denominator.

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
    
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

summary of the data frame


```python
dataframe.info()
```

Conclusion - There is no NULL value in dataframe.



Type of Resturant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

Dining
Conclusion: The majority of the restaurants fall into the dining category.
Dining restaurants are preferred by a larger number of individuals.


```python
grouped_data = dataframe.groupby('listed_in(type)') ['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="black", marker="o")
plt.xlabel("Type of restaurant", c = "black", size = 20)
plt.ylabel("Votes", c = "black", size = 20)
```

The majority of restaurants received ratings

```python
plt.hist(dataframe['rate'], bins=5)
plt.title("Ratings Distribution")
plt.show()
```

Conclusion: The majority of restaurants received ratings ranging from 3.5 to 4.


```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

The majority of couples prefer restaurants with an approximate cost of 300 rupees.

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data =  dataframe)
```

CONCLUSION: Offline orders received lower ratings in comparison to online orders, which obtained excellent ratings.

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order' , aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

CONCLUSION: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants, but prefer online ordering at cafes.
